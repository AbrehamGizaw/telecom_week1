```python
import pandas as pd

# Step 1: Read the CSV Data
df = pd.read_csv('../telecom_data.csv')

# Step 2: Identify Top 10 Handsets
top_10_handsets = df['Handset Type'].value_counts().head(10)
print("Top 10 Handsets:")
print(top_10_handsets)

# Step 3: Identify Top 3 Handset Manufacturers
df['Manufacturer'] = df['Handset Type'].str.split().str[0]
top_3_manufacturers = df['Manufacturer'].value_counts().head(3)
print("Top 3 Handset Manufacturers:")
print(top_3_manufacturers)

# Step 4: Identify Top 5 Handsets per Manufacturer
top_5_handsets_per_manufacturer = {}
for manufacturer in top_3_manufacturers.index:
    top_5_handsets_per_manufacturer[manufacturer] = df[df['Manufacturer'] == manufacturer]['Handset Type'].value_counts().head(5)
print("Top 5 Handsets per Manufacturer:")
for manufacturer, top_5_handsets in top_5_handsets_per_manufacturer.items():
    print(f"{manufacturer}:")
    print(top_5_handsets)

# Step 5: Aggregate Information per User
user_aggregate = df.groupby('MSISDN/Number').agg({
    'Bearer Id': 'count',
    'Dur. (ms)': 'sum',
    'Total DL (Bytes)': 'sum',
    'Total UL (Bytes)': 'sum',
    'Social Media DL (Bytes)': 'sum',
    'Social Media UL (Bytes)': 'sum',
    'Google DL (Bytes)': 'sum',
    'Google UL (Bytes)': 'sum',
    'Email DL (Bytes)': 'sum',
    'Email UL (Bytes)': 'sum',
    'Youtube DL (Bytes)': 'sum',
    'Youtube UL (Bytes)': 'sum',
    'Netflix DL (Bytes)': 'sum',
    'Netflix UL (Bytes)': 'sum',
    'Gaming DL (Bytes)': 'sum',
    'Gaming UL (Bytes)': 'sum',
    'Other DL (Bytes)': 'sum',
    'Other UL (Bytes)': 'sum'
})
print("Aggregate Information per User:")
print(user_aggregate.head())


# Step 6: Interpretation and Recommendations
# Provide insights and recommendations based on the analysis

# Save the results to a new CSV file or display them as needed
```
